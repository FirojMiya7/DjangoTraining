# Control Statement

```python
# if condition:
#           # code to execute if condition is true
#   # else:
#          # code to execute if condition is false
```

```python
marks = int(input("Enter your marks: "))
print("You entered:", marks)
if marks >= 80:
    print("A+")
elif marks >= 70:
    print("A")
elif marks >= 60:
    print("B")
elif marks >= 50:
    print("C")
elif marks >= 40:
    print("D")
else:
    print("Failed")
```

##### Nested if statement

```python
age = int(input("Enter your age: "))
if age >= 18:
    print("You are eligible to vote.")
    if age >= 60:
        print("You are a senior citizen.")
    else:
        print("You are an adult.")
```

##### Swapping two numbers without using temporary variable

# Loops in python

```python
# While Condition
# Repeats the code while the condition is true

i = 1
while i <= 5:
    print(i)
    i += 1
```

##### Range Function

```python
# range(n) generates a sequence of numbers from 0 to n-1
for i in range(5):
    print(i)
```

```python
# range(start, stop) generates a sequence of numbers from start to stop-1
for i in range(1, 6):
    print(i)
```

```python
# range(start, stop, step) generates a sequence of numbers from start to stop-1 with a step value
for i in range(1, 11, 2):
    print(i)
```

##### For Loop

```python
for i in range(1,6):
    print(i)
```

```python
for i in range(1, 21, 3):
    print(i)
```

##### Loop with String

```python
# Loop through a character string
for char in "Hello":
    print(char)
```

```python
# Loop through a list of strings
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

```python
# Break and Continue Statements

# break exit in the loop completely
for i in range(1, 11):
    if i == 5:
        break
    print(i)
```

```python
# Continue skips that iteration and continue for other iterations
for i in range(1, 8):
    if i == 3:
        continue
    print(i)
```

```python
# for else
# The else block is executed when the loop completes normally (i.e., without a break)
for i in range(1, 6):
    print(i)
else:
    print("Loop completed successfully")
```

```python
# nested loops
for i in range(1, 4):
    for j in range(1, 4):
        print(i, j, sep=" ")
        # print(i, j, end=" ")
```

```python
# Multiplication Table

num = int(input("Enter a number to print its multiplication table: "))
for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
```

```python
# Factorial of number 

num = int(input("Enter a number to calculate its factorial: "))
facto = 1
for i in range(1, num + 1):
    facto *= i
print(f"The factorial of {num} is {facto}")
```

```python
# Sum of number from 1 to n

num = int(input("Enter a number to calculate the sum from 1 to n: "))
sum = 0
for i in range(1, num + 1):
    sum += i
print(f"The sum of numbers from 1 to {num} is {sum}")
```

```python
# Reserve a string

string = input("Enter a string to reverse: ")
rev = ""
for char in string:
    rev = char + rev
print(f"The reverse of the string '{string}' is '{rev}'")
```

```python
# Prime number check

num = int(input("Enter a number: "))
if num > 1:
    for i in range(2, num):     # how it works: for example, if num is 7, it checks divisibility by 2, 3, 4, 5, 6
        if (num % i) == 0:
            print(f"{num} is not a prime number.")
            break
    else:
        print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

```python
# Count vowels in a string

string = input("Enter a string: ")
vowels = "aeiouAEIOU"
count = 0
for char in string:
    if char in vowels:
        count += 1
        print(f"Found vowel: {char}")
        print(f"Current count of vowels: {count}")
print(f"\nThe total number of vowels in the string '{string}' is {count}.")
```

```python
# Find largest number in a list

numbers = [10, 20, 5, 40, 15]
largest = numbers[0]  # Assume the first number is the largest
for num in numbers:
    if num > largest:
        largest = num
print(f"The largest number in the list {numbers} is {largest}.")
```

```python
# String indexing and slicing

s = "Python programming"

print("Original string:", s)
print("First character:", s[0])  # Output: P
print("Last character:", s[-1])  # Output: g
print("Substring (0 to 5):", s[0:6])  # Output: Python
print("Substring (7 to end):", s[7:])  # Output: programming
print("Substring (2 to 5):", s[2:5])  # Output: tho
print("Substring (start to end with step 2):", s[::2])  # Output: Pto rgamn
print("Substring (start to end with step -1):", s[::-1])  # Output: gnimmargorp nohtyP
```

```python
# Membership [in , not in]

s = "Java programming"

print("Java" in s)  # Output: True
print("Python" not in s)  # Output: True
print("programming" in s)  # Output: True
print("C++" not in s)  # Output: True
print("Java" not in s)  # Output: False
print("Apple" in s)  # Output: False
print("Newton" not in s)  # Output: True
```

```python
# String method

s = "Python programming with django"

print("Original string:", s)

print("Uppercase:", s.upper())  # Output: PYTHON PROGRAMMING WITH DJANGO
print("Lowercase:", s.lower())  # Output: python programming with django
print("Title Case:", s.title())  # Output: Python Programming With Django
print("Count of 'o':", s.count('o'))  # Output: 3
print("Find 'programming':", s.find('programming'))  # Output: 7
print(s.replace("django", "fastapi"))  # Output: Python programming with fastapi
print("Split by space:", s.split())  # Output: ['Python', 'programming', 'with', 'django']
print("_".join(["I", "love", "to", "code"]))  # Output: I_love_to_code
```
