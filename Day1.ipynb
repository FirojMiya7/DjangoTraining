```python
# Prime number check

num = int(input("Enter a number: "))
if num > 1:
    for i in range(2, num):     # how it works: for example, if num is 7, it checks divisibility by 2, 3, 4, 5, 6
        if (num % i) == 0:
            print(f"{num} is not a prime number.")
            break
    else:
        print(f"{num} is a prime number.")
else:
    print(f"{num} is not a prime number.")
```

# Dictionaries

```python
# Dictionaries store the data as a key value pair
# It is defined under the curly braces {}

student = {
    "name": "Firoj",
    "age": 22,
    "address": "Butwal"
}
print(student)
print(student["name"])
```

```python
# Accessing the dictionary items 
print(student.get("age"))  # Using get() method to access the value associated with the key "age"
print(student.get("gender", "Not Specified"))  # returns default if not available in the dictionary
```

```python
# Adding new element or updating the dictionary

student = {
    "name": "Firoj",
    "age": 22,
    "address": "Butwal"
}

student["gender"] = "Male"  
student["name"] = "Firoj Miya"
student["id"] = 123
print(student)
```

```python
# Pop in dictionary

student.pop("id", "Not Found")  # Removes the key "id" and its associated value from the dictionary
print(student)
```

```python
# Del in dictionary

del student["address"]  # Removes the key "address" and its associated value from the dictionary
print(student)
```

```python
# To remove all items from the dictionary.
student.clear()  # Removes all items from the dictionary
print(student)
```

```python
# Dictionary methods

shoes = {
    "brand": "Nike",
    "size": 10,
    "color": "black",
    "price": 8000
}

print(shoes.keys())  
print(shoes.values()) 
print(shoes.items())  
```

```python
# Nested dictionaries

students = {
    "student1": {
        "name": "Firoj",
        "age": 22,
        "address": "Butwal"
    },
    "student2": {
        "name": "Miya",
        "age": 20,
        "address": "Palpa"
    }
}

print(students.values())
print(students.keys())
print(students.items())
print(students["student1"]["name"])  # Output: Firoj
print(students["student2"])
```

```python
# Merge to dictionaries
dict1 = {
    "a": 1, 
    "b": 2
    }
dict2 = {
    "c": 3,
    "d": 4
    }

merged_dict = {**dict1, **dict2}    # Merges dict1 and dict2 into a new dictionary
print(merged_dict)  # Output: {'a': 1, 'b': 2, 'c': 3, 'd': 4}
```

```python
dict1 = {
    "name": "Firoj",
    "age": 22
}
dict2 = {
    "id": 123,
    "address": "Butwal"
}

dict1.update(dict2)  # Updates dict1 with the key-value pairs from dict2
print("Updated dictionary Successfully:", dict1)  # Output: {'name': 'Firoj', 'age': 22, 'id': 123, 'address': 'Butwal'}
```

```python
# Lambda function (anonymous functions)
# It is a one line functions.

square = lambda x: x * x

print(square(5))  # Output: 25
print()

add = lambda x, y: x + y
print(add(3, 5))  # Output: 8
```

```python
# Create a function with *args that prints the average of the number

def print_average(*args):
    if args:
        average = sum(args) / len(args)
        print(f"The average of {args} is: {average}")
    else:
        print("No numbers provided.")

print_average(10, 15, 30)  
print_average()
```

```python
# Create a lambda function that checks whether the number is +ve, -ve or zero

check_number = lambda x : "positive" if x > 0 else ("negative" if x < 0 else "zero")

print(check_number(10))  
print(check_number(-5))  
print(check_number(0))
```

```python
# Create a function is_prime() that checks whether a number is prime or not
# or not, return true if prime and false otherwise

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

print(is_prime(10))  
print(is_prime(7))
```

```python
# Write a function to print reverse of a string 

def reverse_string(s):
    return s[::-1]

print(reverse_string("Mango"))  
```
